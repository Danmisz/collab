#Ajustando o modelo árvore de decisão

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
seed = 123
X_treino, X_teste, y_treino, y_teste = train_test_split(X,
                                                        y,
                                                        test_size = 0.25,
                                                        stratify = y,
                                                        random_state = seed)
model = DecisionTreeClassifier(max_depth=3, random_state = seed)
#model.fit(X_treino, y_treino)
```

```python
X_treino_novo, X_val, y_treino_novo, y_val = train_test_split(X_treino, y_treino, test_size=0.30, random_state=seed, stratify=y_treino)  #validação, prever
model.fit(X_treino_novo, y_treino_novo)
```

```python
print(model.score(X_val, y_val))
```

```python
model.predict(X_treino_novo)
```

```python
def media_desvio(dados, metrica):
  media = dados[metrica].mean()
  devio_padrao = round(dados[metrica].std(),4)
  # print(f"{metrica}: {media} +- {devio_padrao}")
  print(f"{metrica} média: {round(media - 2*devio_padrao,3)} -{round(media + 2*devio_padrao,3)}")
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_validate, KFold
from sklearn.metrics import accuracy_score,precision_score,recall_score, f1_score
def intervalo_conf(resultados):
    media = resultados['test_score'].mean()
    desvio_padrao = resultados['test_score'].std()
    print(f'Intervalo de confiança: [{media - 2*desvio_padrao}, {min(media + 2*desvio_padrao, 1)}]')
```

###Métrica de avaliação

```python
#metricas
previsoes = model.predict(X_val)
print(f"Acurácia: {accuracy_score(y_val, previsoes)}")
print(f"Recall: {recall_score(y_val, previsoes)}")
print(f"Precision: {precision_score(y_val, previsoes)}")
print(f"F1: {f1_score(y_val, previsoes)}")
```

```python
from sklearn.metrics import confusion_matrix
matriz = confusion_matrix(y_val, previsoes)
print(matriz)              #Como um caso de falso negativo é o mais preocupante se tratando do câncer de mama, a métrica escolhida foi o recall
```

image.png

```python
from sklearn.metrics import classification_report
print(classification_report(y_val, previsoes))
```

###Balanceamento e validação cruzada

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=seed)

x_balanceado, y_balanceado = sm.fit_resample(X_treino_novo, y_treino_novo)
```

```python
skf = StratifiedKFold(n_splits = 5, shuffle = True, random_state = seed)
cv_resultados = cross_validate(model, x_balanceado, y_balanceado, cv = skf, scoring = 'recall')
intervalo_conf(cv_resultados)
```

#Grid Search

```python
from sklearn.model_selection import GridSearchCV
param_grid_dt = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [2,3,6,8,10,12],
}
```

```python
dt_grid_search =  GridSearchCV(estimator = DecisionTreeClassifier(random_state = seed),
                  param_grid=param_grid_dt,
                  scoring = "recall",
                  n_jobs = -1,
                  cv = skf)

dt_grid_search.fit(x_balanceado, y_balanceado)
```

#Pipeline

```python
from imblearn.pipeline import Pipeline as imbpipeline
model = DecisionTreeClassifier(random_state=seed)
pipeline = imbpipeline([('oversample', SMOTE()), ('arvore', model)])
```

```python
skf = StratifiedKFold(n_splits = 5, shuffle = True, random_state = seed)
cv_resultados = cross_validate(pipeline, X_treino_novo, y_treino_novo, cv = skf, scoring = 'recall')
intervalo_conf(cv_resultados)
```

```python
param_grid_dt = {
    'arvore__criterion': ['gini', 'entropy'],
    'arvore__max_depth': [2,3,6,8,10,12],
}
```

```python
dt_grid_search = GridSearchCV(estimator=pipeline,
                param_grid=param_grid_dt,
                scoring = "recall",
                n_jobs = -1,
                cv = skf)

dt_grid_search.fit(X_treino_novo, y_treino_novo)
```

```python
dt_grid_search.best_params_
```

#Grid Search aninhado

```python
inner_cv = StratifiedKFold(shuffle = True, random_state = seed)
outer_cv = StratifiedKFold(n_splits = 3, shuffle = True, random_state = seed)
```

```python
from sklearn.model_selection import cross_val_score

dt_nested_gs = GridSearchCV(estimator=pipeline,
                param_grid=param_grid_dt,
                scoring = "recall",
                n_jobs = -1,
                cv = inner_cv)

dt_nested_scores = cross_val_score(dt_nested_gs, X_treino_novo, y_treino_novo, cv=outer_cv)
```

```python
print(f'resultado de cada iteração do cv externo: {dt_nested_scores}')
print(f'média: {dt_nested_scores.mean()}')
```

```python
dt_nested_gs.fit(X_treino_novo, y_treino_novo)
dt_nested_gs.best_params_
```

#Comparação das métricas

```python
y_pred_val = dt_nested_gs.predict(X_val)  #Comparação com dados de validação
print(confusion_matrix(y_val, y_pred_val))
```

```python
print(classification_report(y_val, y_pred_val))
```
